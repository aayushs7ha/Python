```python
import os
import pandas as pd

# Create an empty list to store DataFrames from CSVs
dataframes = []

# Specify the directory where your CSV files are located
csv_directory = 'C:\\Users\\sinha.ttl\\Desktop\\Jan'

# Loop through all CSV files in the directory
for filename in os.listdir(csv_directory):
    if filename.endswith(".csv"):
        file_date = filename[:-4]  # Remove the '.csv' extension
        df = pd.read_csv(os.path.join(csv_directory, filename))
        df['Date'] = file_date  # Add a new column with the date from the filename
        df.drop(columns='Date',inplace=True)
        dataframes.append(df)

# Concatenate all DataFrames into one
Jan_data = pd.concat(dataframes, ignore_index=True)

# Now you have a single DataFrame with the added 'Date' column
#print(merged_df)
```

```python
Jan_data.to_parquet("January.parquet")
```
