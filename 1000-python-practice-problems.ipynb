# 1. Find the second highest number in a list
- Steps define a function(assign a variable to be passed in the function)
- convert the list into a set to remove duplicates if any
- sort it 
- retrieve second last element 

```python
def second_highest(numbers):
    numbers = list(set(numbers))
    numbers.sort()
    return numbers[-2]

numbers = [100,200,312,1223,411,90,123,234,241,53,87]
print("Second Highest Number : -",second_highest(numbers))
```

# 2. Write a function to compute a factorial of a number using recursion 

```python
def factorial(num):
    if num == 0:
        return 1 
    return num * factorial(num - 1)
num = 5
print(f'Factorial of {num} is :- ',factorial(num))
```

# 3. Reverse a List: Write a function to reverse a given list.

```python
def reverseString(str):
    if str == '':
        print('Nothing 2 return')
        return str
    return str[::-1]
var2 = 'python'
print(f'Reverse of {var2} is :',reverseString(var2))
```

# 4. Find the Maximum and Minimum Element in a List: Write a function to find the maximum and minimum elements in a list.

```python
def MinMax(lst):
    if not lst:
        return 'Empty list'
    # Initialize min and max with the first element of the list
    min_element = lst[0]
    max_element = lst[0]
    for item in lst:
        if isinstance(item, (int, float)):
            if item < min_element:
                min_element = item
            if item > max_element:
                max_element = item
        else:
            return 'Invalid list elements, Non-numeric list'
    return min_element,max_element

list_a = [10,20,30,40,12,31,8,7,11]
min_element, max_element = MinMax(list_a)
print("Minimum element of the given list is ",min_element)
print("Maximum element of the given list is ",max_element)
```

# 5. Sum of Elements in a List: Write a function to calculate the sum of all elements in a list.

```python
def sum_of_elements(lst):
    if not lst:
        return 'Empty List'
    total_sum = 0
    for item in lst:
        if isinstance(item,(int,float)):
            total_sum += item
        else:
            return 'Invalid list'
    return total_sum

x = [1,2,3,4,5,6,7,8,9]        
print("Sum of elements:", sum_of_elements(x))
# Test with an empty list
print("Sum of elements:", sum_of_elements([]))

# Test with a list containing non-numeric values
print("Sum of elements:", sum_of_elements([1, 'a', 3, 4]))
```
